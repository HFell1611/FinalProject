Selecting Dataset, importing and describing

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/HFell1611/FinalProject/main/survey%20lung%20cancer.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
print(df.shape)
```

```python
df.columns
```

```python
feature = ['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']
```

```python
x = df.loc[:, feature]
y = df.loc[:, 'LUNG_CANCER']

y = y.iloc[x.index]
print(x.shape)
print(y.shape)
x.head()
```

```python
convert = lambda d: 1 if d=='M' else 0

x_enc = x['GENDER'].map(convert)
```

```python
print(x_enc.shape)
```

```python
x_df = pd.concat([x_enc.rename('GENDER ENC'), x], axis=1)
```

```python
x_df.drop(['GENDER'], axis=1)
```

```python
convert = lambda d: 1 if d=='YES' else 0

y_enc = y.map(convert)
```
